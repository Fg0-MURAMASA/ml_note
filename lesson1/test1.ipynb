```python
import os

os.makedirs(os.path.join('..', 'data'), exist_ok=True)
data_file = os.path.join('..', 'data', 'house_tiny.csv')
with open(data_file, 'w') as f:
    f.write('NumRooms,Alley,Price\n')
    f.write('NA,Pave,127500\n')
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')
    # python 文件操作, 可以看到真的创建了一个data文件夹. 里面有一个house_tiny.csv文件
```

```python
# 我们通常使用pandas来进行数据处理
import pandas as pd

data = pd.read_csv(data_file)
print(data)
```

```python
# 另一种形式, 直接输出data
data
```

```python
# 未知数据的处理
 # input为前两列的值, output为第三列的值
# fillna是指填补NaN的数据, 用input.mean填补, input.mean是已知数据的平均值
inputs, outputs = data.iloc[:, 0:1], data.iloc[:, 3]
inputs = inputs.fillna(inputs.mean()) # 新版pandas较为严格, 不允许不可求平均值的数据求平均值
print(inputs)
```

```python
# 对类别值或离散值的处理, 我们单独将NaN视为一个类别
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2]
inputs = pd.get_dummies(inputs, dummy_na=True) #这段函数将NaN单独分出来了
inputs = inputs.fillna(inputs.mean())
inputs
```

```python
# 经过上面的操作, 我们不仅处理了所有NaN值, 还将所有的值转化为了数值类型, 这样就方便我们将其转化为张量了
# pandas对象中数据类型是混合的或不是NumPy可以直接处理的类型（如字符串或Pandas特有的数据类型），则可能不会按预期工作
# 所以这里用to_numpy先转换成numpy数组格式
x, y = torch.tensor(inputs.to_numpy(dtype=float)), torch.tensor(outputs.to_numpy(dtype=float))
x, y
```
